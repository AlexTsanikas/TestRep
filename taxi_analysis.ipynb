```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


taxi = sns.load_dataset('taxis')  # Der Datensatz hei√üt 'taxis' in seaborn[1][2]

taxi.to_csv('taxi.csv', index=False)


plt.figure(figsize=(8, 5))
sns.boxplot(x=taxi['total'])
plt.title('Boxplot der Spalte "total"')
plt.show()
```
